```python
# generate C3
accX = df.filter(regex='accX', axis=1).to_numpy()
accX_max = np.amax(accX, axis=1)
```

```python
accY = df.filter(regex='accY', axis=1).to_numpy()
accY_max = np.amax(accY, axis=1)
```

```python
accZ = df.filter(regex='accZ', axis=1).to_numpy()
accZ_max = np.amax(accZ, axis=1)
```

```python
accX_min = np.amin(accX, axis=1)
accY_min = np.amin(accY, axis=1)
accZ_min = np.amin(accZ, axis=1)
```

```python
accX_peak = accX_max - accX_min
accY_peak = accY_max - accY_min
accZ_peak = accZ_max - accZ_min
```

```python
accX_peak[:10]
```

```python
accY_peak[:10]
```

```python
accZ_peak[:10]
```

```python
C3_acc = np.sqrt(accX_peak**2 + accY_peak**2 + accZ_peak**2)
```

```python
# generate C1
accR = df.filter(regex='Resultant', axis=1).to_numpy()
C1_acc = np.sqrt(np.sum(accR**2, axis=1))
```

```python
# generate C2
accXY = df.filter(regex='acc(X|Y)', axis=1).to_numpy()
C2_acc = np.sqrt(np.sum(accXY**2, axis=1))
```

```python
# generate C4
angleZ = np.sqrt(np.sum(accX, axis=1)**2 + np.sum(accZ, axis=1)**2)
vertical = -np.sum(accY, axis=1)
C4_acc = np.arctan2(angleZ, vertical)
```

```python
gyr_df = df.filter(regex='gyr', axis=1)
gyr_df.head(5)
```

```python
gyr = gyr_df.to_numpy()
gyr_r = []

for i in range(len(gyr)):
    elem = []
    for j in range(window_size):
        elem.append(np.linalg.norm(gyr[i, j*3:j*3 + 3]))
    gyr_r.append(elem)

gyrR = np.array(gyr_r)
```

```python
gyr_min = np.amin(gyr_r, axis=1)
```

```python
gyr_max = np.amax(gyr_r, axis=1)
```

```python
# generate C3
gyrX = df.filter(regex='gyrX', axis=1).to_numpy()
gyrX_max = np.amax(gyrX, axis=1)
```

```python
gyrY = df.filter(regex='gyrY', axis=1).to_numpy()
gyrY_max = np.amax(gyrY, axis=1)
```

```python
gyrZ = df.filter(regex='gyrZ', axis=1).to_numpy()
gyrZ_max = np.amax(gyrZ, axis=1)
```

```python
gyrX_min = np.amin(gyrX, axis=1)
gyrY_min = np.amin(gyrY, axis=1)
gyrZ_min = np.amin(gyrZ, axis=1)
```

```python
gyrX_peak = gyrX_max - gyrX_min
gyrY_peak = gyrY_max - gyrY_min
gyrZ_peak = gyrZ_max - gyrZ_min
```

```python
C3_gyr = np.sqrt(gyrX_peak**2 + gyrY_peak**2 + gyrZ_peak**2)
```

```python
# generate C1
C1_gyr = np.sqrt(np.sum(gyrR**2, axis=1))
```

```python
# generate C2
gyrXY = df.filter(regex='gyr(X|Y)', axis=1).to_numpy()
C2_gyr = np.sqrt(np.sum(gyrXY**2, axis=1))
```

```python
# generate C4
angle_gyr_Z = np.sqrt(np.sum(gyrX, axis=1)**2 + np.sum(gyrZ, axis=1)**2)
gyr_vertical = -np.sum(gyrY, axis=1)
C4_gyr = np.arctan2(angle_gyr_Z, gyr_vertical)
```

```python
print(C1_gyr.shape)
print(C2_gyr.shape)
print(C3_gyr.shape)
print(C4_gyr.shape)
```

```python
results = []

for i in range(len(C1_gyr)):
    elem = []
    for attr in [C1_acc, C2_acc, C3_acc, C4_acc, C1_gyr, C2_gyr, C3_gyr, C4_gyr]:
        elem.append(attr[i])
    results.append(elem)

results = np.array(results)
```

```python
new_df = pd.DataFrame(results, columns = ['C1_acc', 'C2_acc', 'C3_acc', 'C4_acc', 'C1_gyr', 'C2_gyr', 'C3_gyr', 'C4_gyr'])
```
